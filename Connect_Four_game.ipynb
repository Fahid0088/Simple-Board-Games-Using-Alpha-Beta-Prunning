```python
def print_grid (grid):
    print ('<---------------(Connect_Four_Game)--------------->'.center(130))
    for row in grid:
        str = ''
        for cell in row:
            str = str + cell
            str = str + '       '
        print (str.center(136))

    
        print ()
        print ()
    print ('C0      C1      C2      C3      C4      C5      C6'.center(130))

def is_winner (player, grid):
    

    for row in grid:
        for i in range (0,4):
            list1 = []
            for j in range (i, i+4):
                list1.append(row[j])
            if list1 == [player, player, player, player]:
                return True
    

    for k in range (7):
        for i in range (0 , 3):
            list1 = []
            for j in range (i, i+ 4):
                list1.append(grid[j][k])

            if list1 == [player, player, player, player]:
                return True
    
    for i in range (3):
        for j in range (4):
            list1 = []
            for k in range (4):
                list1.append(grid[i+k][j+k])
            if list1 == [player, player, player, player]:
                return True
    
    for i in range (3):
        for j in range (3,7):
            list1 = []
            for k in range (4):
                list1.append(grid[i+k][j-k])
            if list1 == [player, player, player, player]:
                return True
            
    return False

def is_draw (grid) :

    if is_winner('X', grid) or is_winner('O', grid)  :
        return False
    for row in grid:
        if '-' in row:
            return False
    return True

def get_possible_moves(grid):
    positions = []
    for col in range(7):
        for row in range(5,-1,-1):
            if grid[row][col] == '-':
                positions.append((row,col))
                break

    return positions

def Alpha_Beta (grid, alpha, beta, depth, is_max):

    if is_winner('X', grid):
        return 1
    if is_winner ('O', grid):
        return -1
    if is_draw(grid) or depth == 0:
        return 0
    
    positions = get_possible_moves(grid)

    if is_max:
        best_score = -10000
        for pos in positions:
            x,y = pos
            grid[x][y] = 'X'
            score = Alpha_Beta (grid, alpha, beta, depth - 1, False)
            grid[x][y] = '-'

            if score > best_score:
                best_score = score

            if score > alpha:
                alpha = score

            if alpha >= beta:
                break
        return best_score
    else:
        best_score = 10000
        for pos in positions:
            x,y = pos
            grid[x][y] = 'O'
            score = Alpha_Beta (grid, alpha, beta, depth - 1, True)
            grid[x][y] = '-'

            if score < best_score:
                best_score = score

            if score < beta:
                beta = score

            if alpha >= beta:
                break
        return best_score

def get_row(col, grid):
    
    for row in range(5,-1,-1):
        if grid[row][col] == '-':
            return row
    return None

def get_best_move(grid):
    best_score = -100000
    best_mov = None

    positions = get_possible_moves(grid)

    for pos in positions:
        x,y  = pos
        grid[x][y] = 'X'
        score = Alpha_Beta (grid, -10000, 10000, 6, False)
        grid[x][y] = '-'

        if score > best_score:
            best_score = score
            best_mov = pos
    
    return best_mov
```

```python
grid = [['-' for i in range(7)] for j in range(6)]


player1 = False

while True:
    print_grid(grid)
    if is_winner('X', grid):
        print('AI Win')
        break
    if is_winner('O', grid):
        print('You Win')
        break
    if is_draw( grid):
        print('Draw')
        break

    if player1:
        x,y = get_best_move(grid)
        grid[x][y] = 'X'
        player1 = False
    else:
        while (True):
            col = int (input ('Enter Valid Column Number:'))
            row = get_row (col, grid)
            if row is not None:

                grid[row][col] = 'O'
                player1 = True
                break
```
