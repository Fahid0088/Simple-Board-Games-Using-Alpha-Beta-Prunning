```python
def print_grid (grid):
    print ('<---------------(Ultimate Tic Tac Toe Game)--------------->'.center(136))
    print()
    print ('-------------------------------------'.center(136))
    for j,row in enumerate(grid):
        str = ''
        for i, cell in enumerate(row):
            if i == 0:
                str = str + ' | '
            str = str + cell
            if  i%3 == 2:
                str = str + ' | '
            else:
                str = str + '   '
        print (str.center(136))
        if  j%3 == 2:
            print ('-------------------------------------'.center(136))
        
def check_sub_grid_win (player, pos, grid):

    x,y = pos
    for i in range (x*3, x*3 + 3):
        list1 = []
        for j in range(y*3, y*3 + 3):
            list1.append(grid[i][j])
        
        if list1 == [player, player, player]:
            return True

    for i in range (y*3, y*3 + 3):
        list1 = []
        for j in range(x*3, x*3 + 3):
            list1.append(grid[j][i])
        
        if list1 == [player, player, player]:
            return True
    
    row = x*3
    col = y*3

    list1 = []
    for i in range (3):
        list1.append(grid[row+i][col+i])

    if list1 == [player, player, player]:
        return True
    
    row = x*3
    col = y*3 + 3 -1

    list1 = []
    for i in range (3):
        list1.append(grid[row+i][col-i])

    if list1 == [player, player, player]:
        return True
    
    return False

def is_winner(player, grid):
    list1 = []
    for i in range(3):
        list2= []
        for j in range(3):
            list2.append(check_sub_grid_win(player, (i,j) , grid))
        list1.append(list2)
    
    for row in list1:
        if row == [True, True, True]:
            return True
    
    for i in range(3):
        col = [row[i] for row in list1]
        if col == [True, True, True]:
            return True

    diag1 = []
    diag2 = []

    for i in range(3):
        for j in range(3):
            if (i==j):
                diag1.append(list1[i][j])
            if (i+j == 2):
                diag2.append(list1[i][j])
    
    if diag1 == [True, True, True]:
        return True
    if diag2 == [True, True, True]:
        return True
    
    return False

def check_sub_grid_draw ( pos, grid):

    if check_sub_grid_win ('X', pos, grid) or check_sub_grid_win ('O', pos, grid):
        return False

    x,y = pos
    for i in range (x*3, x*3 + 3):
        for j in range(y*3, y*3 + 3):
            if grid[i][j] == '-':
                return False
    
    return True

def check_draw (grid):
    
    for i in range (3):
        for j in range (3):
            if (not check_sub_grid_draw ((i,j), grid)) and (not check_sub_grid_win ('X', (i,j), grid)) and (not check_sub_grid_win ('O', (i,j), grid)):
                return False
    return True


def get_possible_moves (grid, current_pos):

    positions = []
    if  check_sub_grid_draw (current_pos, grid) or  check_sub_grid_win ('X', current_pos, grid) or  check_sub_grid_win ('O', current_pos, grid) or (current_pos is None):
        for i in range(9):
            for j in range (9):
                if grid[i][j] == '-':
                    positions.append((i,j))
        return positions
    
    positions = []
    
    x,y = current_pos
    for i in range (x*3, x*3 + 3):
        for j in range(y*3, y*3 + 3):
            if grid[i][j] == '-':
                positions.append((i,j))
    return positions


def Alpha_Beta (grid, alpha, beta, depth, is_max, current_pos):

    if is_winner('X', grid):
        return 1
    if is_winner ('O', grid):
        return -1
    if check_draw(grid) or depth == 0:
        return 0
    
    positions = get_possible_moves(grid, current_pos)

    if is_max:
        best_score = -10000
        for pos in positions:
            x,y = pos
            grid[x][y] = 'X'
            score = Alpha_Beta (grid, alpha, beta, depth - 1, False, (x%3, y%3))
            grid[x][y] = '-'

            if score > best_score:
                best_score = score

            if score > alpha:
                alpha = score

            if alpha >= beta:
                break
        return best_score
    else:
        best_score = 10000
        for pos in positions:
            x,y = pos
            grid[x][y] = 'O'
            score = Alpha_Beta (grid, alpha, beta, depth - 1, True, (x%3, y%3))
            grid[x][y] = '-'

            if score < best_score:
                best_score = score

            if score < beta:
                beta = score

            if alpha >= beta:
                break
        return best_score


def get_best_move(grid, current_pos):
    best_score = -100000
    best_mov = None

    positions = get_possible_moves(grid, current_pos)

    for pos in positions:
        x,y  = pos
        grid[x][y] = 'X'
        score = Alpha_Beta (grid, -10000, 10000, 2, False, (x%3, y%3))
        grid[x][y] = '-'

        if score > best_score:
            best_score = score
            best_mov = pos
    
    return best_mov
```
